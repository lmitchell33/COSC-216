```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

url = 'https://raw.githubusercontent.com/lmitchell33/COSC-216/main/Cancer_Data.csv'
df = pd.read_csv(url)
df.info() # shows the organized data frame 
```

```python
print(df.isnull().sum()) # no null values in the data set
df = df.drop(labels='Unnamed: 32', axis=1)
df.info()
```

```python
sns.countplot(x='diagnosis', data=df)
plt.title('Number of Cells Recorded')
plt.xlabel('Malignant or Benign')
plt.ylabel('Number of Cells')
# more benign than malignant
```

```python
malignant  = df[df.diagnosis == 'M']
benign = df[df.diagnosis == 'B']
plt.scatter(malignant.radius_mean, malignant.texture_mean, label = "Malignant", marker = 's', alpha = 0.5)
plt.scatter(benign.radius_mean, benign.texture_mean, label = "Benign", alpha = 0.5)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title("Radius vs. Texture")
plt.legend()
plt.show()
```

```python
plt.scatter(malignant.radius_mean, malignant.area_mean, label = "Malignant")
plt.scatter(benign.radius_mean, benign.area_mean, label = "Benign")
plt.xlabel('Mean Radius')
plt.ylabel('Mean Area')
plt.title('Radius vs. Area')
plt.legend()
plt.show()
```

Assuming the cell is shapes like that of a sphere/cylinder, the area would just be a funciotn of the radius to a certain power, therefore, the graph makes sense

```python
plt.scatter(malignant.perimeter_mean, malignant.radius_mean, label = "Malignant")
plt.scatter(benign.perimeter_mean, benign.radius_mean, label = "Benign")
plt.xlabel('Mean Perimeter')
plt.ylabel('Mean Radius')
plt.title('Perimeter vs. Radius')
plt.legend()
plt.show()
```

If the cell is larger, there is a greater radius and therefore, a greater perimeter

```python
plt.scatter(malignant.perimeter_mean, malignant.area_mean, label = "Malignant")
plt.scatter(benign.perimeter_mean, benign.area_mean, label = "Benign")
plt.xlabel('Mean Perimeter')
plt.ylabel('Mean Area')
plt.title('Perimeter vs. Area')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x=malignant.compactness_mean, y=malignant.concavity_mean, data = df, label='Malignant')
sns.scatterplot(x=benign.compactness_mean, y=benign.concavity_mean, data=df, label='Benign')
plt.xlabel('Mean Compactness')
plt.ylabel('Mean Concavity')
plt.title('Compactness vs. Concavity')
plt.legend()
plt.show()
```

With more compactness = more concavity ?? Not too sure about this but it is linear so that is what the data depicts

```python
sns.scatterplot(x=malignant['concave points_mean'], y=malignant.concavity_mean, label = 'Malignant')
sns.scatterplot(x=benign['concave points_mean'], y=benign.concavity_mean, label='Bengin')
plt.xlabel('Mean Concave Points')
plt.ylabel('Mean Concavity')
plt.title('Concave Points vs Concavity')
plt.legend()
plt.show()
```

```python
smoothness = [malignant['smoothness_mean'], benign['smoothness_mean']]
sns.boxplot(data=smoothness)
plt.xlabel("Cacner Type")
plt.ylabel("Smoothness Value")
plt.title("Malignant vs. Benign Smoothness")
```

```python
symmetry = [malignant.symmetry_mean, benign.symmetry_mean]
sns.boxplot(data=symmetry)
plt.xlabel("Cacner Type")
plt.ylabel("Symmetry Value")
plt.title("Malignant vs. Benign Symmetry")
```

```python
compactness = [malignant.compactness_mean, benign.compactness_mean]
sns.boxplot(data=compactness)
plt.xlabel("Cacner Type")
plt.ylabel("Compactness Value")
plt.title("Malignant vs. Benign Compactness")
```

```python
texture = [malignant.texture_mean, benign.texture_mean]
sns.boxplot(data=texture)
plt.xlabel("Cacner Type")
plt.ylabel("Texture Value")
plt.title("Malignant vs. Benign Texture")
```

```python
FD = [malignant.fractal_dimension_mean, benign.fractal_dimension_mean]
sns.boxplot(data=FD)
plt.xlabel("Cacner Type")
plt.ylabel("FD Value")
plt.title("Malignant vs. Benign Fractal Dimension")
```

```python
names = ['radius_mean', 'texture_mean', 'perimeter_mean','area_mean','smoothness_mean','compactness_mean',
         'concavity_mean','concave points_mean','symmetry_mean', 'fractal_dimension_mean']
meanMalignant = {}
meanBenign = {}
for i in names:
  meanMalignant[i] = malignant[i].mean()
  meanBenign[i] = benign[i].mean()
mean_both = {}
for index in meanMalignant:
  mean_both[index] = [meanMalignant[index], meanBenign[index]]
mean_df = pd.DataFrame(mean_both)
mean_df = mean_df.rename(index={0:'Malignant', 1:'Benign'})
mean_df = mean_df.T
print(mean_df)
```

```python
# Standard Error Values
SE_names = ['radius_se', 'texture_se', 'perimeter_se','area_se','smoothness_se','compactness_se',
         'concavity_se','concave points_se','symmetry_se', 'fractal_dimension_se']
malignantSE = {}
benignSE = {}
for i in SE_names:
  malignantSE[i] = malignant[i].mean()
  benignSE[i] = benign[i].mean()
SE_both = {}
for index in malignantSE:
  SE_both[index] = [malignantSE[index], benignSE[index]]
SE_df = pd.DataFrame(SE_both)
SE_df = SE_df.rename(index={0:'Malignant', 1:'Benign'})
SE_df = SE_df.T
print(SE_df)
```
